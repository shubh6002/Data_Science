<h1 align='center' style='color:green'>Clustering With K Means </h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("income.csv")
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1['Age'], df1['Income($)'], color = 'green')
plt.scatter(df2['Age'], df2['Income($)'], color = 'red')
plt.scatter(df3['Age'], df3['Income($)'], color = 'black')

plt.xlabel("Age")
plt.ylabel('Income($)')
```

```python
# green points are not properly cluster
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df[['Income($)']] = scaler.transform(df[['Income($)']])
df
# income coloum changes to o - 1 scale
```

```python
# Doing the same thing for Age

scaler.fit(df[['Age']])
df[['Age']] = scaler.transform(df[['Age']])
df
```

```python
df
```

```python
# Training our scale data
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['Age', 'Income($)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
# thess clusters are the centroids
km.cluster_centers_
```

```python
df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]

plt.scatter(df1['Age'], df1['Income($)'], color = 'green')
plt.scatter(df2['Age'], df2['Income($)'], color = 'red')
plt.scatter(df3['Age'], df3['Income($)'], color = 'black')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], color = 'purple', marker= '*', label= 'Centroid')

plt.xlabel("Age")
plt.ylabel('Income($)')
plt.legend()
```

#### Elbow plot technique

The above problem is simple with 3 features , but in real life whatif there are more than 20 features, for that elbow plot technique will be usefull.

```python
k_rng = range(1,10)
sse = []     #sse = sum of square error
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age', 'Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('k')
plt.ylabel('Sum of square error')
plt.plot(k_rng, sse)
```
